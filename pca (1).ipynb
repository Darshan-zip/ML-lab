```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, roc_auc_score, roc_curve
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier, StackingClassifier
from xgboost import XGBClassifier
import matplotlib.pyplot as plt
import seaborn as sns

# ----------------------------
# Step 1: Dataset loading and description
# ----------------------------
def load_dataset():
    # Example: UCI Breast Cancer Wisconsin (Diagnostic) dataset
    url = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data"
    col_names = ['ID', 'Diagnosis'] + [f'feature_{i}' for i in range(1, 31)]
    df = pd.read_csv(url, header=None, names=col_names)
    return df

# Display dataset description & distribution
def dataset_description(df):
    print("Dataset shape:", df.shape)
    print("Target classes distribution:\n", df['Diagnosis'].value_counts())
    print("Sample of features:\n", df.head())
    
# ----------------------------
# Step 2: Preprocessing (standardization, encoding, imputation)
# ----------------------------
def preprocess_data(df):
    df = df.copy()
    # Drop ID column
    df.drop(columns=['ID'], inplace=True)
    
    # Encode Diagnosis target column ('M'->1, 'B'->0)
    le = LabelEncoder()
    df['Diagnosis'] = le.fit_transform(df['Diagnosis'])
    
    # Separate features and target
    X = df.drop(columns=['Diagnosis'])
    y = df['Diagnosis']
    
    # Impute missing values (if any)
    imputer = SimpleImputer(strategy='mean')
    X_imputed = imputer.fit_transform(X)
    
    # Standardize features
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X_imputed)
    
    return X_scaled, y, scaler

# ----------------------------
# Step 3: Apply PCA with variance target and plot scree plot
# ----------------------------
def apply_pca(X, variance_target=0.95):
    pca = PCA(n_components=variance_target, random_state=42)
    X_pca = pca.fit_transform(X)
    print(f"Number of components chosen to explain {variance_target*100}% variance: {pca.n_components_}")
    
    # Scree Plot
    plt.figure(figsize=(10,6))
    plt.plot(np.cumsum(pca.explained_variance_ratio_)*100)
    plt.xlabel('Number of Components')
    plt.ylabel('Cumulative Explained Variance (%)')
    plt.title('PCA Scree Plot')
    plt.grid(True)
    plt.show()
    
    return X_pca, pca

# ----------------------------
# Step 4: Define hyperparameter grids for models
# ----------------------------
param_grids = {
    'SVM': {
        'kernel': ['linear', 'rbf'],
        'C': [0.1, 1, 10],
        'gamma': ['scale', 'auto']
    },
    'NaiveBayes': {
        'var_smoothing': [1e-9, 1e-8, 1e-7]
    },
    'KNN': {
        'n_neighbors': [3, 5, 7],
        'weights': ['uniform', 'distance'],
        'metric': ['euclidean', 'manhattan']
    },
    'LogisticRegression': {
        'penalty': ['l2'],
        'C': [0.1, 1, 10],
        'solver': ['liblinear']
    },
    'DecisionTree': {
        'criterion': ['gini', 'entropy'],
        'max_depth': [None, 5, 10]
    },
    'RandomForest': {
        'n_estimators': [100, 200],
        'max_depth': [None, 5, 10],
        'criterion': ['gini', 'entropy']
    },
    'AdaBoost': {
        'n_estimators': [50, 100],
        'learning_rate': [0.01, 0.1, 1]
    },
    'GradientBoosting': {
        'n_estimators': [100, 200],
        'learning_rate': [0.01, 0.1],
        'max_depth': [3, 5]
    },
    'XGBoost': {
        'n_estimators': [100, 200],
        'learning_rate': [0.01, 0.1],
        'max_depth': [3, 5]
    }
}

# ----------------------------
# Step 5: Define models
# ----------------------------
def get_model(name):
    if name == 'SVM':
        return SVC(probability=True, random_state=42)
    elif name == 'NaiveBayes':
        return GaussianNB()
    elif name == 'KNN':
        return KNeighborsClassifier()
    elif name == 'LogisticRegression':
        return LogisticRegression(random_state=42)
    elif name == 'DecisionTree':
        return DecisionTreeClassifier(random_state=42)
    elif name == 'RandomForest':
        return RandomForestClassifier(random_state=42)
    elif name == 'AdaBoost':
        return AdaBoostClassifier(random_state=42)
    elif name == 'GradientBoosting':
        return GradientBoostingClassifier(random_state=42)
    elif name == 'XGBoost':
        return XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
    else:
        raise ValueError(f"Model {name} not recognized.")

# ----------------------------
# Step 6: Perform 5-Fold cross-validation with hyperparameter tuning and collect fold-wise performance
# ----------------------------
def cross_validate_model(X, y, model_name, param_grid, cv_folds=5):
    model = get_model(model_name)
    skf = StratifiedKFold(n_splits=cv_folds, shuffle=True, random_state=42)
    
    acc_scores = []
    best_params_list = []
    
    fold = 1
    for train_index, val_index in skf.split(X, y):
        X_train, X_val = X[train_index], X[val_index]
        y_train, y_val = y.iloc[train_index], y.iloc[val_index]
        
        grid_search = GridSearchCV(model, param_grid, cv=3, scoring='accuracy', n_jobs=-1)
        grid_search.fit(X_train, y_train)
        
        best_model = grid_search.best_estimator_
        y_pred = best_model.predict(X_val)
        acc = accuracy_score(y_val, y_pred)
        acc_scores.append(acc)
        best_params_list.append(grid_search.best_params_)
        
        print(f"{model_name} - Fold {fold} Accuracy: {acc:.4f}")
        fold += 1
    
    avg_acc = np.mean(acc_scores)
    print(f"{model_name} Average Accuracy over {cv_folds} folds: {avg_acc:.4f}")
    return acc_scores, avg_acc, best_params_list

# ---------------------------
# Step 7: Evaluate all models no-PCA and with-PCA
# ---------------------------
def run_experiment():
    df = load_dataset()
    dataset_description(df)

    X_scaled, y, scaler = preprocess_data(df)
    
    # PCA application with 95% variance target
    X_pca, pca = apply_pca(X_scaled, variance_target=0.95)
    
    models = ['SVM', 'NaiveBayes', 'KNN', 'LogisticRegression', 'DecisionTree',
              'RandomForest', 'AdaBoost', 'GradientBoosting', 'XGBoost']
    
    results = {'Model': [], 'PCA': [], 'Fold Accuracies': [], 'Average Accuracy': [], 'Best Params': []}
    
    print("\n### Evaluating models WITHOUT PCA ###")
    for model_name in models:
        accs, avg_acc, best_params = cross_validate_model(X_scaled, y, model_name, param_grids[model_name])
        results['Model'].append(model_name)
        results['PCA'].append(False)
        results['Fold Accuracies'].append(accs)
        results['Average Accuracy'].append(avg_acc)
        results['Best Params'].append(best_params)
    
    print("\n### Evaluating models WITH PCA ###")
    for model_name in models:
        accs, avg_acc, best_params = cross_validate_model(X_pca, y, model_name, param_grids[model_name])
        results['Model'].append(model_name)
        results['PCA'].append(True)
        results['Fold Accuracies'].append(accs)
        results['Average Accuracy'].append(avg_acc)
        results['Best Params'].append(best_params)
    
    results_df = pd.DataFrame(results)
    
    # Comparison pivot table of average accuracy
    pivot = results_df.pivot(index='Model', columns='PCA', values='Average Accuracy')
    print("\nAverage Accuracy comparison (No-PCA vs With-PCA):")
    print(pivot)
    
    return results_df, pca

# ---------------------------
# Step 8: Plot confusion matrix and ROC curve for best performing models
# ---------------------------
def plot_confusion_roc(X_train, y_train, X_test, y_test, model_name, param_grid):
    model = get_model(model_name)
    grid_search = GridSearchCV(model, param_grid, cv=3, scoring='accuracy', n_jobs=-1)
    grid_search.fit(X_train, y_train)
    best_model = grid_search.best_estimator_
    
    y_pred = best_model.predict(X_test)
    y_proba = None
    if hasattr(best_model, 'predict_proba'):
        y_proba = best_model.predict_proba(X_test)[:, 1]
    
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix - {model_name}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
    
    if y_proba is not None:
        fpr, tpr, _ = roc_curve(y_test, y_proba)
        roc_auc = roc_auc_score(y_test, y_proba)
        plt.plot(fpr, tpr, label=f'ROC AUC = {roc_auc:.4f}')
        plt.plot([0, 1], [0, 1], linestyle='--')
        plt.title(f'ROC Curve - {model_name}')
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.legend()
        plt.show()

if __name__ == "__main__":
    # Run experiment and get results
    results_df, pca = run_experiment()
    
    # For demonstration, split and plot confusion matrix / ROC for SVM (no PCA)
    df = load_dataset()
    X_scaled, y, scaler = preprocess_data(df)
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, stratify=y, test_size=0.2, random_state=42)
    print("\nConfusion matrix and ROC for SVM without PCA:")
    plot_confusion_roc(X_train, y_train, X_test, y_test, 'SVM', param_grids['SVM'])
    
    # Similarly can plot for other models or PCA data
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

def plot_confusion_matrix_pca(df, model_name, param_grid, variance_target=0.95):
    # Preprocess dataset
    X_scaled, y, _ = preprocess_data(df)
    
    # Split original dataset
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, stratify=y, test_size=0.2, random_state=42)
    
    # Apply PCA on training and test sets
    pca = PCA(n_components=variance_target, random_state=42)
    X_train_pca = pca.fit_transform(X_train)
    X_test_pca = pca.transform(X_test)
    
    # Define model and perform hyperparameter tuning
    model = get_model(model_name)
    grid_search = GridSearchCV(model, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
    grid_search.fit(X_train_pca, y_train)
    best_model = grid_search.best_estimator_
    
    # Predict on test data
    y_pred = best_model.predict(X_test_pca)
    
    # Compute confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    
    # Plot confusion matrix
    plt.figure(figsize=(6,5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix for {model_name} with PCA')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.show()

# Example to use for SVM (replace model_name and param_grid for other models)
df = load_dataset()
plot_confusion_matrix_pca(df, 'SVM', param_grids['SVM'])
```

```python
from sklearn.metrics import roc_curve, roc_auc_score

def plot_roc_curve_pca(df, model_name, param_grid, variance_target=0.95):
    # Preprocess dataset
    X_scaled, y, _ = preprocess_data(df)

    # Split original dataset
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, stratify=y, test_size=0.2, random_state=42)

    # Apply PCA on training and test sets
    pca = PCA(n_components=variance_target, random_state=42)
    X_train_pca = pca.fit_transform(X_train)
    X_test_pca = pca.transform(X_test)

    # Define model and perform hyperparameter tuning
    model = get_model(model_name)
    grid_search = GridSearchCV(model, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
    grid_search.fit(X_train_pca, y_train)
    best_model = grid_search.best_estimator_

    # Predict probabilities on test data (for positive class)
    if hasattr(best_model, "predict_proba"):
        y_proba = best_model.predict_proba(X_test_pca)[:, 1]
    else:
        # For models without predict_proba (e.g. SVM with probability=False), use decision_function
        y_proba = best_model.decision_function(X_test_pca)
    
    # Compute ROC curve and AUC
    fpr, tpr, _ = roc_curve(y_test, y_proba)
    roc_auc = roc_auc_score(y_test, y_proba)

    # Plot ROC curve
    plt.figure(figsize=(7,6))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.3f})')
    plt.plot([0,1], [0,1], color='navy', lw=2, linestyle='--')
    plt.title(f'ROC Curve for {model_name} with PCA')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend(loc='lower right')
    plt.grid(True)
    plt.show()

# Example usage for SVM with PCA
df = load_dataset()
plot_roc_curve_pca(df, 'SVM', param_grids['SVM'])
```
